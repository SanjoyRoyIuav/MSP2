Import libraries

```python
import pandas as pd
path = "https://github.com/SanjoyRoyIuav/MSP2/blob/main/Total_Conflict.xlsx?raw=true"
toc = pd.read_excel(path)
print(toc.head(3))
```

```python
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

plt.figure(figsize=(2.5, 1.2), dpi=300) #defining figure size and resolution

Labels=["IW","TS","CZ","EEZ",'CS'] #X axis label ordering

sns.set(font_scale = 0.7) #x and y axis tick label size control

flierprops = dict(marker='o', markerfacecolor='None', markersize=1,  markeredgecolor='black') #changing outlier markers size and shape
meanprops = dict(marker='o', markerfacecolor='white', markersize=2,  markeredgecolor='red') #changing outlier markers size and shape


sns.boxplot(x="Sea", y="Conflict_score", data=toc, order=Labels, showmeans=True, meanprops=meanprops, flierprops=flierprops, linewidth=0.5, palette="deep", )
plt.ylabel("Conflict score", size= 11)
plt.xlabel("Ocean planning units", size= 11)

plt.show()
```
